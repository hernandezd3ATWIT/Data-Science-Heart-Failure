# Early Mortality Risk in Heart-Failure Patients

This data science project analyzes clinical data from heart failure patients to identify key factors associated with early mortality. Using demographic, lifestyle, and lab test variables, we explore how different conditions influence survival rates. The project also builds a predictive model to assist clinicians in quickly assessing patient risk.

## Research Questions

1. **Demographic Impact**: How do age and sex affect the probability of a death event?
2. **Comorbidities & Lifestyle**: Do diabetes, anaemia, high blood pressure, or smoking significantly increase mortality risk?
3. **Lab Biomarkers**: Which clinical measures (e.g., ejection fraction, creatinine) are most predictive of survival?
4. **Predictive Modeling**: Can we predict a patientâ€™s death event using features collected at admission?

## Dataset
- Source: [Kaggle - Heart Failure Clinical Records Dataset](https://www.kaggle.com/datasets/andrewmvd/heart-failure-clinical-data)
- 299 patient records
- 13 clinical and demographic features

## Tools & Libraries
- Python (Pandas, Matplotlib, Seaborn, Scikit-learn)
- Jupyter Notebook


```python
# -----------------------------------------------------------
# Q1: Demographic Impact - Age and Sex vs Death Event
# -----------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (make sure it's in the same folder as your notebook)
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
# -----------------------------------------------------------
# Plot 1: Age vs Death Event (Larger Boxplot)
# -----------------------------------------------------------
plt.figure(figsize=(10, 7))  # Bigger figure
sns.boxplot(data=df, x='DEATH_EVENT', y='age')
plt.title("Age vs Death Event", fontsize=16)
plt.xlabel("Death Event (0 = Survived, 1 = Died)", fontsize=13)
plt.ylabel("Age (years)", fontsize=13)
plt.xticks([0, 1], ['Survived', 'Died'], fontsize=12)
plt.yticks(fontsize=11)
plt.tight_layout()
# plt.savefig("Q1_Age_vs_DeathEvent.png", dpi=300)
plt.show()


# -----------------------------------------------------------
# Plot 2: Sex vs Death Event (Larger + Legend Fixed)
# -----------------------------------------------------------
sex_death_pct = pd.crosstab(df['sex'], df['DEATH_EVENT'], normalize='index') * 100
sex_death_pct.columns = ['Survived', 'Died']
sex_death_pct.index = ['Female', 'Male']

plt.figure(figsize=(10, 7))  # Bigger figure
ax = sex_death_pct.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], edgecolor='black', figsize=(10, 7))
plt.title("Sex vs Patient Outcome", fontsize=16)
plt.xlabel("Sex", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=14)
plt.xticks(rotation=0, fontsize=13)
plt.yticks(fontsize=12)

# Move legend to top-right OUTSIDE the plot
plt.legend(title="Patient Outcome", title_fontsize=12, fontsize=11, loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.tight_layout()
# plt.savefig("Q1_Sex_vs_DeathEvent.png", dpi=300, bbox_inches='tight')  # Tight layout to include legend
plt.show()
```

```python
features = ['anaemia', 'diabetes', 'high_blood_pressure', 'smoking']

for feature in features:
    # Create percentage breakdown
    ctab = pd.crosstab(df[feature], df['DEATH_EVENT'], normalize='index') * 100
    ctab.columns = ['Survived', 'Died']
    ctab.index = ['No', 'Yes']

    # Plot
    ax = ctab.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'],
                   edgecolor='black', figsize=(8, 6))
    ax.set_title(f"{feature.replace('_', ' ').title()} vs Death Event", fontsize=15)
    ax.set_xlabel(f"{feature.replace('_', ' ').title()}", fontsize=13)
    ax.set_ylabel("Percentage (%)", fontsize=13)
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=11)
    ax.legend(title="Patient Outcome", title_fontsize=12)
    plt.xticks(rotation=0)

    # Save the figure correctly using its own figure object
    fig = ax.get_figure()
    fig.tight_layout()
    fig.savefig(f"Q2_{feature}_vs_DeathEvent.png", dpi=300, bbox_inches='tight')

    plt.show()
```
